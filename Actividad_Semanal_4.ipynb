```python
imput = "https://raw.githubusercontent.com/PosgradoMNA/actividades-del-projecto-equipo_112/main/default%20of%20credit%20card%20clients.csv"
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(imput, index_col='ID')
df.rename(columns={'X1': 'Credit', 'X2': 'Gender', 'X3': 'Education', 'X4': 'Marital Status', 'X5': 'Age', 'X6': 'Payments September', 'X7': 'Payments August', 'X8': 'Payments July', 'X9': 'Payments June', 'X10': 'Payments May', 'X11': 'Payments April'}, inplace=True)
df.rename(columns={'X12': 'Bill September', 'X13': 'Bill August', 'X14': 'Bill July', 'X15': 'Bill June', 'X16': 'Bill May', 'X17': 'Bill April'}, inplace=True)
df.rename(columns={'X18': 'Previuos September', 'X19': 'Previuos August', 'X20': 'Previuos July', 'X21': 'Previuos June', 'X22': 'Previous May', 'X23': 'Previuos April'}, inplace=True)
df.head() # mostrar los primeros de los datos
```

```python
df.tail() # mostramos los ultimos registros de filas
```

```python
df.info() # imprime la información de columna y filas
```

```python
nuevo_df = df.copy()
nuevo_df # Copiamos la base de datos para no modificar el original.
```

```python
# Identificar valores nulos
nuevo_df.isnull().sum().sort_values(ascending=True) # Mostrar la suma del conteo de valores nulos
```

```python
# Removemos los valores nulos ya que solo quitariamos el .05% de valores a anlizar
nuevo_df.dropna(axis = 0, inplace=True)

# Corroboramos que no existan valores nulos
nuevo_df.isnull().sum().sort_values(ascending=True)
```

```python
# Solo analizaremos las columnas con valores relacionados a pagos
df_num = nuevo_df.iloc[:,5:11]

# Mostramos la correlacion entre las columnas
df_num.corr()
```

```python
corrs = df_num.corr()
sns.set(rc = {'figure.figsize':(15,10)})
sns.heatmap(corrs, vmin = -1, vmax = 1, cmap = "BuGn", annot= True)
```

```python
 # Buscamos la varianza total de todas las variables
t = df_num.var().sum()

var_payment_september = df_num["Payments September"].var()
var_payment_august = df_num["Payments August"].var()
var_payment_july = df_num["Payments July"].var()

print('Varianza Total \t\t:',t.round(2) )
print('Varianza Pagos de Septiembre \t:', ((var_payment_september/t)*100).round(2),'%' )
print('Varianza Pagos de Agosto \t:', ((var_payment_august/t)*100).round(2),'%' )
print('Varianza Pagos de Julio \t:', ((var_payment_july/t)*100).round(2),'%' )
```

```python
scaler = StandardScaler()

# Escalamos los datos usando StandardScaler

scaled = scaler.fit_transform(df_num)

scaled_df = pd.DataFrame(scaled, columns=df_num.columns)

scaled_df.head()
```

```python
# Confimamos que los datos esten escalados

cols = ["Payments September", "Payments August", "Payments July"]

summary = {
    "Varianzas": scaled_df[cols].var()[0:3],
    "Valor Min": scaled_df[cols].min()[0:3],
    "Valor Max": scaled_df[cols].max()[0:3]
}

pd.DataFrame(summary, index = cols).transpose()
```

```python
# Empieza el PCA

pcs = PCA()

pcs_t = pcs.fit_transform(scaled_df)

pcsSummary_df = pd.DataFrame({
                            '% varianza explicada': np.round(pcs.explained_variance_ratio_,4) * 100,
                            '% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_) * 100 })
```

```python
# Mostramos % de la varianza y acumulado de la varianza de cada componente

pcs_labels = [f'PC{i + 1}' for i in range(len(scaled_df.columns))]
pcsSummary_df.index = pcs_labels
pcsSummary_df
```

```python
# Comprobamos que los componentes mantienen 100% de la información

pcs_df = pd.DataFrame(pcs_t, columns =pcs_labels)
print("Varianza total variables originales: ", scaled_df.var().sum())
print("Varianza total de los componentes: ", pcs_df.var().sum())
```

```python
# Determinamos los componentes principales (En este caso podemos determinar que el PC1, PC2 ya que explican el 80% de la variación)

pcsSummary_df
```

```python
# Mostramos los pesos de cada uno de las variables con los componentes principales

comps_df = pd.DataFrame(pcs.components_.round(4),columns = pcs_df.columns,index = scaled_df.columns)

comps_df.iloc[:,:2]
```

```python
# Obtenemos los valores maximos de cada componente

comps_df.iloc[:,:2].abs().idxmax()
```
