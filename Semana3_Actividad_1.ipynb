```python
# Import libraries

import pandas as pd
import numpy as np
```

```python
# Load data source and represent data value
# Limpiamos y pusimos los nombres correctos a las columnas y tambien la respesantacion de algunos valores

ds = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv', index_col=0)
ds.rename(columns={'X1': 'Credit', 'X2': 'Gender', 'X3': 'Education', 'X4': 'Marital Status', 'X5': 'Age', 'X6': 'Payments September', 'X7': 'Payments August', 'X8': 'Payments July', 'X9': 'Payments June', 'X10': 'Payments May', 'X11': 'Payments April'}, inplace=True)
ds.rename(columns={'X12': 'Bill September', 'X13': 'Bill August', 'X14': 'Bill July', 'X15': 'Bill June', 'X16': 'Bill May', 'X17': 'Bill April'}, inplace=True)
ds.rename(columns={'X18': 'Previuos September', 'X19': 'Previuos August', 'X20': 'Previuos July', 'X21': 'Previuos June', 'X22': 'Bill May', 'X23': 'Previuos April'}, inplace=True)
ds['Gender'] = ds['Gender'].map({
    1: 'Male',
    2: 'Female'
})
ds['Education'] = ds['Education'].map({
    1: 'Graduate',
    2: 'University',
    3: 'Hight School',
    4: 'Other',
})
ds['Marital Status'] = ds['Marital Status'].map({
    1: 'Married',
    2: 'Single',
    3: 'Other',
})
ds['Age'] = ds['Age'].fillna(0).astype('int')
ds
```

```python
# Verifies null values
# Detectamos las columnas que tienes valores en nulo

ds.isnull().values.any()
ds.isnull().any()
```

```python
# Verifies empty values
# Detectamos las columnas que tienes valores en NA

ds.isna().values.any()
ds.isna().any()
```

```python
# Drop columns with empty values
# Pintamos el data frame solo considerando las columna "sucias"

ds1 = ds.copy()
ds1.dropna(axis = 1, inplace = True)
ds1
```

```python
# Drop rows with at least 8 empty values
# Detectamos que solo existen dos registros que no contienen valores en las columnas "Age" y "Credit"

ds3 = ds.copy()
ds3.dropna(thresh=8, inplace = True, axis = 0)
ds3
```

```python
# Remove rows where all values are missing
# No se elimina ninguna fila aqui, ya que ninguna fila tiene todos los valores en nulo, ya se sabia esto por el analisis de las columnas con valores vacios.

ds4 = ds.copy()
ds4.dropna(how='all', inplace = True)
ds4 

# Zero rows deleted
```

```python
# Evaluamos que valor utilizaremos para los reemplazos de "Credit"
ds["Credit"].describe()
```

```python
# Sacamos la media de la columna Credit para poder asignarla a los valores vacios

ndf = ds.copy()
mc = ndf["Credit"].mean()
"MEAN Credit: " + str(mc)
```

```python
# Asignamos la media de la columna Credit a los valores vacios

ndf['Credit'] = ndf['Credit'].fillna(value = mc, inplace = True)
ndf
```

```python
# Evaluamos que valor utilizaremos para los reemplazos de "Age"
ds["Age"].describe()
```

```python
# Sacamos la mediana de la columna Age para poder asignarla a los valores vacios

ma = ndf["Age"].median()
"Median Age: " + str(ma)
```

```python
# Asignamos la mediana de la columna Age a los valores vacios

ndf['Age'] = ndf['Age'].fillna(value = ma, inplace = True)
ndf
```

```python
# Corroboramos que nuestros valores ya esten limpios de valores nulos
ndf.isnull().any()
```

```python
# Corroboramos que nuestros valores ya esten limpios de valores NA
ndf.isna().any()
```
